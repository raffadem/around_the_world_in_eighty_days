```python
# latitude -90,90
# longitude -180,180
def calculate_lat_lng(lat, lng, x_size=0.2, y_size=0.1):
    lat_max=lat+y_size/2
    lat_min=lat-y_size/2
    lng_min=lng
    lng_max=lng+x_size
    
    if lat_max >= 90:
        lat_max -= 180
    
    if lat_min <= -90:
        lat_min += 180
        
    if lng_max >= 180:
        lng_max -= 360
    
    if lng_min <= -180:
        lng_min += 360
    
    print(f"lat: {lat}")
    print(f"\tlat_min: {lat_min}, lat_max: {lat_max}")
    print(f"lng: {lng}")
    print(f"\tlng_min: {lng_min}, lng_max: {lng_max}")
    return lat_min, lat_max, lng_min, lng_max

def query(lat_min, lat_max, lng_min, lng_max):
    if (lat_min >= 0 and lat_max>=0) or (lat_min <= 0 and lat_max<=0):
        lat_condition = f"(lat >= {lat_min} & lat <= {lat_max})"
    elif lat_min >=0 and lat_max<=0:
        lat_condition = f"(lat >= {lat_min} & lat <= 90) | (lat >= -90 & lat <= {lat_max})"
    elif lat_min<=0 and lat_max >=0:
        lat_condition = f"(lat >= {lat_min} & lat <= 0) | (lat >= 0 & lat < {lat_max})"
        
    if (lng_min >= 0 and lng_max>=0) or (lng_min <= 0 and lng_max<=0):
        lng_condition = f"(lng >= {lng_min} & lng <= {lng_max})"
    elif lng_min >=0 and lng_max<=0:
        lng_condition = f"(lng >= {lng_min} & lng <= 180) | (lng >= -180 & lng <= {lng_max})"
    elif lng_min<=0 and lat_max >=0:
        lng_condition = f"(lng >= {lng_min} & lng <= 0) | (lng >= 0 & lng < {lng_max})"
        
    print (f"\t\t{lat_condition}")
    print (f"\t\t{lng_condition}")
```

```python
list_of_coordinates = [
    (90, 180), 
    (-90, 180), 
    (-89.95, 179.9), 
    (89.95, -179.9),
    (-90, -180), 
    (-0, -180), 
    (60, 120), 
    (-60, -120)
]
for coordinates in list_of_coordinates:
    lat_min, lat_max, lng_min, lng_max = calculate_lat_lng(lat=coordinates[0], lng=coordinates[1])
    query(lat_min, lat_max, lng_min, lng_max)
    print()
```
